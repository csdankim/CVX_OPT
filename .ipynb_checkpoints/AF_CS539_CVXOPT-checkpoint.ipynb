```python
"""
who is being retweeted (if any)
who is being tweeted at/mentioned (if any)
what hashtags are being used (if any)
"""

def find_retweeted(tweet):
    '''This function will extract the twitter handles of retweed people'''
    return re.findall('(?<=RT\s)(@[A-Za-z]+[A-Za-z0-9-_]+)', tweet)

def find_mentioned(tweet):
    '''This function will extract the twitter handles of people mentioned in the tweet'''
    return re.findall('(?<!RT\s)(@[A-Za-z]+[A-Za-z0-9-_]+)', tweet)  

def find_hashtags(tweet):
    '''This function will extract hashtags'''
    return re.findall('(#[A-Za-z]+[A-Za-z0-9-_]+)', tweet)  
```

## From Text to Vector ##

```python
# take hashtags which appear at least this amount of times
min_appearance = 10
# find popular hashtags - make into python set for efficiency
popular_hashtags_set = set(popular_hashtags[
                           popular_hashtags.counts>=min_appearance
                           ]['hashtag'])
```

```python
# make a new column with only the popular hashtags
hashtags_list_df['popular_hashtags'] = hashtags_list_df.hashtags.apply(
            lambda hashtag_list: [hashtag for hashtag in hashtag_list
                                  if hashtag in popular_hashtags_set])
# drop rows without popular hashtag
popular_hashtags_list_df = hashtags_list_df.loc[
            hashtags_list_df.popular_hashtags.apply(lambda hashtag_list: hashtag_list !=[])]
```

```python
# make new dataframe
hashtag_vector_df = popular_hashtags_list_df.loc[:, ['popular_hashtags']]

for hashtag in popular_hashtags_set:
    # make columns to encode presence of hashtags
    hashtag_vector_df['{}'.format(hashtag)] = hashtag_vector_df.popular_hashtags.apply(
        lambda hashtag_list: int(hashtag in hashtag_list))
```

```python
hashtag_matrix = hashtag_vector_df.drop('popular_hashtags', axis=1)

# calculate the correlation matrix
correlations = hashtag_matrix.corr()

# plot the correlation matrix
plt.figure(figsize=(10,10))
sns.heatmap(correlations,
    cmap='RdBu',
    vmin=-1,
    vmax=1,
    square = True,
    cbar_kws={'label':'correlation'})
plt.show()
```

## Cleaning Unstructured Text Data ##
I would like to know the general things which people are talking about, not who they are talking about or to and not the web links they are sharing.

```python
def remove_links(tweet):
    '''Takes a string and removes web links from it'''
    tweet = re.sub(r'http\S+', '', tweet) # remove http links
    tweet = re.sub(r'bit.ly/\S+', '', tweet) # rempve bitly links
    tweet = tweet.strip('[link]') # remove [links]
    return tweet

def remove_users(tweet):
    '''Takes a string and removes retweet and @user information'''
    tweet = re.sub('(RT\s@[A-Za-z]+[A-Za-z0-9-_]+)', '', tweet) # remove retweet
    tweet = re.sub('(@[A-Za-z]+[A-Za-z0-9-_]+)', '', tweet) # remove tweeted at
    return tweet
```

```python
my_stopwords = nltk.corpus.stopwords.words('english')
word_rooter = nltk.stem.snowball.PorterStemmer(ignore_stopwords=False).stem
my_punctuation = '!"$%&\'()*+,-./:;<=>?[\\]^_`{|}~•@'

# cleaning master function
def clean_tweet(tweet, bigrams=False):
    tweet = remove_users(tweet)
    tweet = remove_links(tweet)
    tweet = tweet.lower() # lower case
    tweet = re.sub('['+my_punctuation + ']+', ' ', tweet) # strip punctuation
    tweet = re.sub('\s+', ' ', tweet) #remove double spacing
    tweet = re.sub('([0-9]+)', '', tweet) # remove numbers
    tweet_token_list = [word for word in tweet.split(' ')
                            if word not in my_stopwords] # remove stopwords

    tweet_token_list = [word_rooter(word) if '#' not in word else word
                        for word in tweet_token_list] # apply word rooter
    if bigrams:
        tweet_token_list = tweet_token_list+[tweet_token_list[i]+'_'+tweet_token_list[i+1]
                                            for i in range(len(tweet_token_list)-1)]
    tweet = ' '.join(tweet_token_list)
    return tweet
```

```python
df['clean_tweet'] = df.tweet.apply(clean_tweet)
```

```python
df.tail(10)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer(max_df=0.9, min_df=25, token_pattern='\w+|\$[\d\.]+|\S+')

# apply transformation
tfidf = vectorizer.fit_transform(df['clean_tweet']).toarray()

# tf_feature_names tells us what word each column in the matric represents
tfidf_feature_names = vectorizer.get_feature_names() 
print(tfidf)
```

## LDA ##

```python
from sklearn.decomposition import LatentDirichletAllocation

number_of_topics = 10

model_LDA = LatentDirichletAllocation(n_components=number_of_topics, random_state=0)
```

```python
%time model_LDA.fit(tfidf)
```

```python
def display_topics(model, feature_names, no_top_words):
    topic_dict = {}
    for topic_idx, topic in enumerate(model.components_):
        topic_dict["Topic %d words" % (topic_idx+1)]= ['{}'.format(tfidf_feature_names[i])
                        for i in topic.argsort()[:-no_top_words - 1:-1]]
        topic_dict["Topic %d weights" % (topic_idx+1)]= ['{:.3f}'.format(topic[i])
                        for i in topic.argsort()[:-no_top_words - 1:-1]]
    return pd.DataFrame(topic_dict)
```

```python
no_top_words = 10
display_topics(model_LDA, tfidf_feature_names, no_top_words)
```

```python
table_LDA = display_topics(model_LDA, tfidf_feature_names, no_top_words)
table_LDA.to_csv("LDA.csv")
```

## Anchor Free Algorithm ##

```python
from scipy.sparse import spdiags

# ---- NCW procedure ------- #
# References: Wei Xu, Xin Liu, and Yihong Gong.Document clustering Based on Non-negative Matrix Factorization.
#             In Proceedings of the 26th annual international ACM SIGIR conference on Research and developmentin informaion retrieval, 
#             pages 267–273. ACM, 2003.
    # remove zero rows
    # np.where(~tfidf.any(axis=1))[0]
    # a[~(a==0).all(1)]
tfidf = tfidf[~(tfidf==0).all(1)]
D_half = sum(tfidf*tfidf)**(-0.5)
doc_num = D_half.shape[0] 
D_half = spdiags(D_half, 0, doc_num, doc_num)
tfidf = tfidf * D_half

# compute the correlation
D = tfidf
# print(D.shape)
nWords, nDoc = D.shape
P = (1/nDoc) * np.dot(D.T, D)
print(P.shape)  # (425, 425)

F = 10 # number of topics

from scipy.sparse.linalg import eigs
# Applying Eigen decomposition to D matrix

vals, vecs = eigs(P, F)

vals = np.real(vals)
DD = np.diag(vals)
# print(DD)
DD = np.abs(DD)
DD_root = np.sqrt(DD)
B = np.real(np.dot(vecs, DD_root))
# print(B.shape)
# print(B)

# Initialize Identity E
E = np.identity(F)
```

```python
import cvxpy as cp
# def AnchorFree(B, E):
p = B.shape[0]
n = B.shape[1]
M = E
# print(B)
a = np.zeros(n)
mm = 0
conv = []

from tqdm import tqdm

while True:
    for i in tqdm(range(n)):
        for j in range(n):
            step = (-1.)**(i+j)
            N = np.delete(M, j, axis=0)
            N = np.delete(N, i, axis=1)
            Nt = np.linalg.det(N)
            a[j] = (step*Nt)

        x1 = cp.Variable(n)
        obj1 = cp.Maximize(a * x1)
        one = np.ones((p, 1))
        constraints1 = [B * x1 >= 0, (np.dot(one.T, B)) * x1 == 1]
        prob1 = cp.Problem(obj1, constraints1)
        prob1.solve(solver=cp.CVXOPT)
        a1 = prob1.value
        x1 = x1.value

        x2 = cp.Variable(n)
        obj2 = cp.Minimize(a * x2)
        constraints2 = [B * x2 >= 0, (np.dot(one.T, B)) * x2 == 1]
        prob2 = cp.Problem(obj2, constraints2)
        prob2.solve(solver=cp.CVXOPT)
        a2 = prob2.value
        x2 = x2.value

        if abs(a1) > abs(a2):
            M[:, i] = x1;
        else:
            M[:, i] = x2;

        detM = np.linalg.det(M)
        conv.append(abs(detM))
        mm += 1

    # Convergence condition
    idx = len(conv) - 1
    if (mm > 1) and (abs(conv[idx] - conv[idx-1]) < 1e-6):
        break  
```

```python
C = np.dot(B, M)
```

```python
C.shape
```

```python
no_top_words = 10
topic_dict_AF = {}
for k in range(F):
    topic_dict_AF["Topic %d words" % (k+1)]= ['{}'.format(tfidf_feature_names[i]) for i in C[:, k].argsort()[:-no_top_words - 1:-1]]
    topic_dict_AF["Topic %d weights" % (k+1)]= ['{:.4f}'.format(C[:, k][i]) for i in C[:, k].argsort()[:-no_top_words - 1:-1]]    
pd.DataFrame(topic_dict_AF)
```

```python
table_AF = pd.DataFrame(topic_dict_AF)
table_AF.to_csv("AF.csv")
```

## kmeans clustering ##

```python
from sklearn.cluster import KMeans

F = 10 # num of topics
model_kmeans = KMeans(n_clusters=F, init='k-means++', max_iter=100, n_init=1)
%time model_kmeans.fit(tfidf)

order_centroids = model_kmeans.cluster_centers_.argsort()[:, ::-1]
```

```python
no_top_words = 10
topic_dict_kmeans = {}
for k in range(F):
    topic_dict_kmeans["Topic %d words" % (k+1)] = ['{}'.format(tfidf_feature_names[i]) for i in order_centroids[k, :no_top_words]]
    topic_dict_kmeans["Topic %d weights" % (k+1)]= ['{:.4f}'.format(model_kmeans.cluster_centers_.T[:, k][i]) for i in order_centroids[k, :no_top_words]]    
pd.DataFrame(topic_dict_kmeans)
```

```python
table_kmeans = pd.DataFrame(topic_dict_kmeans)
table_kmeans.to_csv("kmeans.csv")
```
